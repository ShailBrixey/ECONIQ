```python
# Import Libraries
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# API Key and Base URL
API_KEY = "fa2afb466f7b43f74136f01fd2cc7c9a"

BASE_URL = "https://api.stlouisfed.org/fred/series/observations"
```

```python
# Federal Funds Rate (FEDFUNDS)
params = {
    "series_id": "FEDFUNDS",
    "api_key": API_KEY,
    "file_type": "json",
    "observation_start": "1970-01-01",
    "observation_end": "2025-01-01"
}

response = requests.get(BASE_URL, params=params)

fedfunds = pd.DataFrame(response.json()["observations"])

fedfunds["date"] = pd.to_datetime(fedfunds["date"])
fedfunds["value"] = pd.to_numeric(
    fedfunds["value"],
    errors="coerce"
)

fedfunds.head()
```

```python
# CPI (CPIAUCSL)
params["series_id"] = "CPIAUCSL"

response = requests.get(BASE_URL, params=params)

cpi = pd.DataFrame(response.json()["observations"])

cpi["date"] = pd.to_datetime(cpi["date"])
cpi["value"] = pd.to_numeric(
    cpi["value"],
    errors="coerce"
)

cpi.head()
```

```python
# Unemployment Rate (UNRATE)
params["series_id"] = "UNRATE"

response = requests.get(BASE_URL, params=params)

unrate = pd.DataFrame(response.json()["observations"])

unrate["date"] = pd.to_datetime(unrate["date"])
unrate["value"] = pd.to_numeric(
    unrate["value"],
    errors="coerce"
)

unrate.head()
```

```python
# Money Supply (M2SL)
params["series_id"] = "M2SL"

response = requests.get(BASE_URL, params=params)

m2 = pd.DataFrame(response.json()["observations"])

m2["date"] = pd.to_datetime(m2["date"])
m2["value"] = pd.to_numeric(
    m2["value"],
    errors="coerce"
)

m2.head()
```

```python
# Industrial Production (INDPRO)
params["series_id"] = "INDPRO"

response = requests.get(BASE_URL, params=params)

indpro = pd.DataFrame(response.json()["observations"])

indpro["date"] = pd.to_datetime(indpro["date"])
indpro["value"] = pd.to_numeric(
    indpro["value"],
    errors="coerce"
)

indpro.head()
```

```python
# Consumer Sentiment (UMCSENT)
params["series_id"] = "UMCSENT"

response = requests.get(BASE_URL, params=params)

sentiment = pd.DataFrame(response.json()["observations"])

sentiment["date"] = pd.to_datetime(sentiment["date"])
sentiment["value"] = pd.to_numeric(
    sentiment["value"],
    errors="coerce"
)

sentiment.head()
```

```python
# Real GDP (GDPC1)
params["series_id"] = "GDPC1"

response = requests.get(BASE_URL, params=params)

gdp = pd.DataFrame(response.json()["observations"])

gdp["date"] = pd.to_datetime(gdp["date"])
gdp["value"] = pd.to_numeric(
    gdp["value"],
    errors="coerce"
)

gdp.head()
```

```python
# 10-Year Treasury Yield (GS10)
params["series_id"] = "GS10"

response = requests.get(BASE_URL, params=params)

gs10 = pd.DataFrame(response.json()["observations"])

gs10["date"] = pd.to_datetime(gs10["date"])
gs10["value"] = pd.to_numeric(
    gs10["value"],
    errors="coerce"
)

gs10.head()
```

```python
# Yield Curve Spread (T10Y2Y)
params["series_id"] = "T10Y2Y"

response = requests.get(BASE_URL, params=params)

yieldcurve = pd.DataFrame(response.json()["observations"])

yieldcurve["date"] = pd.to_datetime(yieldcurve["date"])
yieldcurve["value"] = pd.to_numeric(
    yieldcurve["value"],
    errors="coerce"
)

yieldcurve.head()
```

```python
# Recession Indicator (USREC)
params["series_id"] = "USREC"

response = requests.get(BASE_URL, params=params)

recession = pd.DataFrame(response.json()["observations"])

recession["date"] = pd.to_datetime(recession["date"])
recession["value"] = pd.to_numeric(
    recession["value"],
    errors="coerce"
)

recession.head()
```

```python
# S&P 500 (SP500)
params["series_id"] = "SP500"

response = requests.get(BASE_URL, params=params)

sp500 = pd.DataFrame(response.json()["observations"])

sp500["date"] = pd.to_datetime(sp500["date"])
sp500["value"] = pd.to_numeric(
    sp500["value"],
    errors="coerce"
)

sp500.head()
```

```python
# Rename Columns
fedfunds = fedfunds[["date", "value"]].rename(columns={"value": "FEDFUNDS"})
cpi = cpi[["date", "value"]].rename(columns={"value": "CPIAUCSL"})
unrate = unrate[["date", "value"]].rename(columns={"value": "UNRATE"})
m2 = m2[["date", "value"]].rename(columns={"value": "M2SL"})
indpro = indpro[["date", "value"]].rename(columns={"value": "INDPRO"})
sentiment = sentiment[["date", "value"]].rename(columns={"value": "UMCSENT"})
gdp = gdp[["date", "value"]].rename(columns={"value": "GDPC1"})
gs10 = gs10[["date", "value"]].rename(columns={"value": "GS10"})
yieldcurve = yieldcurve[["date", "value"]].rename(columns={"value": "T10Y2Y"})
recession = recession[["date", "value"]].rename(columns={"value": "USREC"})
sp500 = sp500[["date", "value"]].rename(columns={"value": "SP500"})
```

```python
# Merge All Datasets
macro = (
    fedfunds
    .merge(cpi, on="date", how="outer")
    .merge(unrate, on="date", how="outer")
    .merge(m2, on="date", how="outer")
    .merge(indpro, on="date", how="outer")
    .merge(sentiment, on="date", how="outer")
    .merge(gdp, on="date", how="outer")
    .merge(gs10, on="date", how="outer")
    .merge(yieldcurve, on="date", how="outer")
    .merge(recession, on="date", how="outer")
    .merge(sp500, on="date", how="outer")
)

macro = macro.sort_values("date")

macro.head()
```

```python
# Correlation Matrix
corr = macro.drop(columns=["date"]).corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True)

plt.title("Macroeconomic Correlation Matrix")
plt.show()
```

```python
# Federal Funds Rate Plot
plt.figure(figsize=(12,6))

plt.plot(
    fedfunds["date"],
    fedfunds["FEDFUNDS"]
)

plt.title("Federal Funds Rate")
plt.xlabel("Date")
plt.ylabel("Percent")

plt.show()
```

```python
# CPI Plot
plt.figure(figsize=(12,6))

plt.plot(
    cpi["date"],
    cpi["CPIAUCSL"]
)

plt.title("Consumer Price Index")
plt.xlabel("Date")
plt.ylabel("Index")

plt.show()
```

```python
# Yield Curve vs. Recession
plt.figure(figsize=(14,6))

plt.plot(
    yieldcurve["date"],
    yieldcurve["T10Y2Y"],
    label="10Y-2Y Spread"
)

plt.fill_between(
    recession["date"],
    -5,
    5,
    where=recession["USREC"] == 1,
    alpha=0.3,
    label="Recession"
)

plt.legend()

plt.title("Yield Curve and Recessions")

plt.show()
```
